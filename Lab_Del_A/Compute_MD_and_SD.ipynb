```python
import pandas as pd
import numpy as np
# Read the data into a DataFrame
df = pd.read_csv('Lab_A_Data.csv')

# Extract the numeric part from the 'r_optical' column
df['r_optical_value'] = df['r_optical'].str.extract(r'([\d.]+e[+-]?\d+)').astype(float)

# Split data into group A and group B
df_A = df[df['Row'].str.startswith('A')]
df_B = df[df['Row'].str.startswith('B')]

# Function to calculate and print statistics for a given group
def calculate_stats(group, label):
    mean_D = group['Value D'].mean()
    std_D = group['Value D'].std()

    mean_r_hyd = group['r_hyd'].mean()
    std_r_hyd = group['r_hyd'].std()

    mean_r_optical = group['r_optical_value'].mean()
    std_r_optical = group['r_optical_value'].std()

    print(f"\n{label} Group Stats:")
    print(f"Mean of D: {mean_D:.2e}")
    print(f"Standard Deviation of D: {std_D:.2e}")
    print(f"Mean of r_hyd: {mean_r_hyd:.2e}")
    print(f"Standard Deviation of r_hyd: {std_r_hyd:.2e}")
    print(f"Mean of r_optical: {mean_r_optical:.2e}")
    print(f"Standard Deviation of r_optical: {std_r_optical:.2e}")

# Calculate stats for group A
calculate_stats(df_A, 'Group A')

# Calculate stats for group B
calculate_stats(df_B, 'Group B')
```
