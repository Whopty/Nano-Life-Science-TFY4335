```python
# Load data
import pandas as pd
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
for i in range(1,4):
    df = pd.read_csv(f'optical_radius_A{i}.csv')
    ax.hist(df, bins=50, alpha=0.5, label=f'Dataset A{i}')

ax.set_xlabel('Radius (microns)') 
ax.set_ylabel('Frequency')
ax.set_title('Optical Radius Distribution')
ax.legend()
plt.show()

fig, ax = plt.subplots()
for i in range(1,4):
    df = pd.read_csv(f'optical_radius_B{i}.csv')
    ax.hist(df, bins=50, alpha=0.5, label=f'Dataset B{i}')

ax.set_xlabel('Radius (microns)') 
ax.set_ylabel('Frequency')
ax.set_title('Optical Radius Distribution')
ax.legend()
plt.show()
```
